```python
import tensorflow as tf
import tensorflow_probability as tfp
import tensorflow.math as tm
import numpy as np
import random
import time
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras import initializers
from tensorflow.keras import Model
from tensorflow.keras.layers import Dense
```

```python
x_train = np.array([[0, 0],
           [0, 1],
           [1, 0],
           [1, 1]])
y_train = np.array([[0],
           [1],
           [1],
           [0]])
train_ds = tf.data.Dataset.from_tensor_slices((x_train, y_train)).batch(4)
```

```python
batch_size = 4
epochs = 4000
    
model = keras.Sequential(
        [
            layers.InputLayer(input_shape=(2,)),
            layers.Dense(32, activation = "sigmoid"),
            layers.Dense(1, activation = "sigmoid")
        ]
) 
'''
w_0 = np.array([[1, -1], [1, -1]], dtype = "float32")
b_0 = np.array([-0.5, 1], dtype = "float32")
l_0 = [w_0, b_0]

w_1 = np.array([[1], [1]], dtype = "float32")
b_1 = np.array([-1], dtype = "float32")
l_1 = [w_1, b_1]
        
model.layers[0].set_weights(l_0)
model.layers[1].set_weights(l_1)
'''
opt = tf.keras.optimizers.SGD(learning_rate=0.1)
model.compile(loss="binary_crossentropy", optimizer=opt, metrics=["accuracy"])
history = model.fit(train_ds, batch_size=batch_size, epochs=epochs)
```
